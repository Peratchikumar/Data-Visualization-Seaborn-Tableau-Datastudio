```python
##1.Importing Library
import pandas as pd

#import       → keyword that brings an external library into your code
#pandas       → the library name (data analysis toolkit)
#as           → keyword to give the imported library a short nickname
#pd           → the nickname used to call pandas functions
```

```python
#2.Importing the Dataset
df=pd.read_csv('Zomato Chennai Listing 2020.csv')

#df           → variable name you're creating to hold the data
#=            → assignment operator, stores the right-side result into df
#pd.read_csv  → pandas function that reads a CSV file into a table (DataFrame)
#('Zomato_Chennai_Listing_2020.csv') → the file path/name passed as input to read_csv

#Loads the Chennai dataset into a DataFrame called df
```

```python
#Q1.Which 10 locations in Chennai have the most restaurants listed?
df['Location'].value_counts().head(10)

#df              → the DataFrame holding the dataset
#['Location']    → square brackets select one column by name, here "Location"
#.value_counts() → method that counts how many times each unique value appears
#()              → empty parentheses, no extra settings used
#.head(10)       → method that returns only the first 10 rows of a result
#(10)            → the number passed to head, telling it how many rows to keep

#Shows the top 10 restaurant-dense localities in Chennai
```

```python
#Q2.What is the average "Price for 2" by location, for the top 10 busiest locations?
top_locations = df['Location'].value_counts().head(10).index
df[df['Location'].isin(top_locations)].groupby('Location')['Price for 2'].mean().sort_values(ascending=False)

#top_locations     → new variable name to store the list of top location names
#=                 → assignment operator
#df['Location']    → selects the Location column
#.value_counts()   → counts occurrences of each unique location
#.head(10)         → keeps only the top 10 most frequent
#.index             → attribute that extracts just the labels (location names) from that result, dropping the counts

#df                          → the full DataFrame again
#[df['Location'].isin(top_locations)] → filtering step: keeps only rows where Location is in top_locations
#df['Location']              → the column being checked
#.isin(top_locations)        → method that returns True/False for each row, True if the value matches one in top_locations
#.groupby('Location')        → method that splits the filtered data into groups, one per unique location
#('Location')                → tells groupby which column to group by
#['Price for 2']             → after grouping, selects the Price for 2 column from each group
#.mean()                     → calculates the average value within each group
#.sort_values()              → method to reorder the resulting values
#(ascending=False)           → argument telling sort_values to put the largest numbers first

#Compares average dining cost across Chennai's most popular localities
```

```python
#Q3.Correlation between price and dining rating
df[['Price for 2', 'Dining Rating']].corr()

#df                              → the DataFrame
#[['Price for 2', 'Dining Rating']] → double brackets select multiple columns as a smaller DataFrame
#'Price for 2'                  → first column name in the selection list
#'Dining Rating'                → second column name in the selection list
#.corr()                        → method that calculates the Pearson correlation coefficient between numeric columns

#Tells you whether costlier restaurants tend to have better or worse dining ratings
```

```python
#Q4.Most common cuisines
import ast
from collections import Counter

all_cuisines = df['Cuisine'].dropna().apply(ast.literal_eval).sum()
Counter(all_cuisines).most_common(10)

#import ast                 → brings in Python's "ast" module, used here to safely read string-lists
#from                        → keyword to import only part of a library
#collections                 → built-in Python module containing useful data structures
#import Counter              → imports just the Counter tool from that module

#all_cuisines    → new variable name to store the combined list of cuisines
#=               → assignment operator
#df['Cuisine']   → selects the Cuisine column (currently stored as text, e.g. "['Biryani', 'Chinese']")
#.dropna()       → removes any rows where Cuisine is missing/empty
#.apply(...)     → method that runs a function on every value in the column
#ast.literal_eval → function that converts a string that looks like a list/dict into an actual Python list/dict
#.sum()          → here, since each row becomes a list, sum() joins (concatenates) all those lists into one long list

#Counter(all_cuisines) → function that counts how many times each item appears in the list
#.most_common(10)      → method that returns the 10 most frequently occurring items, with their counts

#Reveals the most popular cuisine types across all Chennai listings
```

```python
#Q5.Biggest gap between dining and delivery ratings
df['Rating Gap'] = df['Dining Rating'] - df['Delivery Rating']
df[['Name of Restaurant', 'Location', 'Dining Rating', 'Delivery Rating', 'Rating Gap']].dropna().sort_values('Rating Gap', ascending=False).head(10)

#df['Rating Gap']      → creates a new column named "Rating Gap" inside df
#=                      → assignment operator, defines what goes into that new column
#df['Dining Rating']    → selects the existing Dining Rating column
#-                      → subtraction operator, applied row by row
#df['Delivery Rating']  → selects the existing Delivery Rating column

#df                                  → the DataFrame (now containing the new Rating Gap column)
#[[...]]                             → double brackets selecting multiple specific columns to display
#'Name of Restaurant'                → column name for restaurant name
#'Location'                          → column name for restaurant location
#'Dining Rating'                     → column name for dine-in rating
#'Delivery Rating'                   → column name for delivery rating
#'Rating Gap'                        → the new column just created
#.dropna()                           → removes rows with any missing values among the selected columns
#.sort_values('Rating Gap', ascending=False) → reorders rows by Rating Gap, largest values first
#.head(10)                           → keeps only the top 10 rows of that sorted result

#Flags restaurants where the dine-in experience and delivery experience differ the most
```
