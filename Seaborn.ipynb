```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing the dataset
dataset=pd.read_csv("PrePlacement.csv")
dataset  #Function call
```

```python
#Number of rows and columns
print("The Number of Row",dataset.shape[0])
print("The Number of Column",dataset.shape[1])
```

```python
dataset.shape
```

```python
#Univariate Analysis
import seaborn as sb
sb.distplot(dataset['ssc_p'],hist=True,kde=True,kde_kws={'color':'red'})	
```

```python
#Bivariate Analysis - Distribution Plot
sb.jointplot(x='ssc_p',	y='salary',data=dataset)
plt.show()
```

```python
sb.jointplot(x='ssc_p',	y='salary',data=dataset,kind='hex')
plt.show()
```

```python
sb.jointplot(x='ssc_p',	y='salary',data=dataset,kind='kde') #KDE - Kernel Density Estimation
plt.show()
```

```python
#Pair Plot
sb.pairplot(dataset,hue='gender',diag_kind="kde",kind="scatter",palette="husl") #Hue - Do Differenciate
plt.show()
```

```python
#For Categorical Plot
sb.stripplot(x="ssc_b",y="ssc_p",data=dataset)
plt.show()            
```

```python
sb.stripplot(x="gender",y="ssc_p",data=dataset,jitter=True)
plt.show()            
```

```python
sb.swarmplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

```python
#Distribution Observation - Important Plot
sb.violinplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

```python
sb.violinplot(x="specialisation",y="salary",data=dataset) #Without hue=gender
plt.show()
```

```python
sb.violinplot(x="specialisation",y="salary",data=dataset,hue='gender') #With hue=gender
plt.show()
```

```python
sb.boxplot(x="ssc_b",y="ssc_p",data=dataset) #Without hue=gender
plt.show()
```

```python
sb.boxplot(x="ssc_b",y="ssc_p",data=dataset,hue='gender') #With hue=gender
plt.show()
```

```python
sb.barplot(x="ssc_b",y="ssc_p",data=dataset) #Without hue=gender
plt.show()
```

```python
sb.barplot(x="ssc_b",y="ssc_p",data=dataset,hue='gender') #With hue=gender
plt.show()
```

```python
sb.countplot(x="ssc_b",data=dataset) #Without hue=gender
plt.show()
```

```python
sb.countplot(x="ssc_b",data=dataset,hue='gender') #With hue=gender
plt.show()
```

```python
sb.pointplot(x="ssc_p",y="hsc_b",data=dataset) #Without hue=gender
plt.show()
```

```python
sb.pointplot(x="ssc_p",y="hsc_b",data=dataset,hue='gender') #With hue=gender
plt.show()
```

```python
sb.boxplot(data=dataset,orient="h")
plt.show()
```

```python
#Catplot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('exercise')
sb.catplot(x="time",y="pulse",hue="kind",kind="violin",col="diet",data=df);
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.regplot(x="total_bill",y="tip",data=df) #regplot - Regression Plot
sb.lmplot(x="total_bill",y="tip",data=df) #lmplot - Linear Model Plot
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('anscombe')
sb.lmplot(x="x",y="y",data=df.query("dataset=='II'"))
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('anscombe')
sb.lmplot(x="x",y="y",data=df.query("dataset=='II'"),order=2)
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```
