```python
#4.Cleaning the dataset
#A.Removing the redundunt rows of data

#Checking redundunt rows of data
wrong_data = raw_df["PAGE NO"] =="PAGE NO"
raw_df[wrong_data]

#1.wrong_data - A variable that stores a True/False (boolean) mask for each row
#2.raw_df["PAGE NO"] - Selects the "PAGE NO" column from the DataFrame
#3.== "PAGE NO" - Checks if the column value equals the text "PAGE NO" — returns True if yes, False if no
#4.raw_df[wrong_data] - Filters and displays only the rows where wrong_data is True

#raw_df[wrong_data] - Identifies repeated header rows mistakenly entered as data, so they can be removed during cleaning
```

```python
#Performing Negation of the wrong dataset and then storing the correct data back in the raw_df DataFrame
#This permanently remove the wrong data from the original dataframe

raw_df=raw_df[~wrong_data] 

#1.raw_df - The DataFrame being updated (overwritten with clean data)
#2.= Assignment — stores the filtered result back into raw_df
#3.raw_df[...] - Filters rows from the DataFrame
#4.~ Negation operator — flips True → False and False → True
#5.wrong_data - The boolean mask where True = junk rows (PAGE NO header rows)
#6.~wrong_data - Now True = good rows, False = junk rows


#~wrong_data - flips the filter to keep only valid rows, permanently removing junk header rows from the DataFrame
```

```python
#Dropping columns which are not required for further analysis
raw_df.drop(['URL','PAGE NO','CITY'],axis=1,inplace=True)

#1.raw_df - The Zomato Mumbai DataFrame
#2..drop() - Method that removes specified rows or columns
#3.['URL', 'PAGE NO', 'CITY'] - List of column names to be deleted
#4.axis=1 - Tells pandas to drop columns (axis=1) not rows (axis=0)
#5.inplace=True - Permanently modifies raw_df without needing reassignment

#Permanently removes 3 unnecessary columns (URL, PAGE NO, CITY) to keep only the columns useful for analysis
```

```python
#Replacing the other null records with NA
raw_df.fillna('NA', inplace=True)

#1.raw_df - The Zomato Mumbai DataFrame
#2..fillna() - Method that fills missing (NaN) values with a specified value
#3.'NA' - The replacement value — fills all NaN cells with the text "NA"
#4.inplace=True - Permanently modifies raw_df without needing reassignment

#Replaces all remaining NaN missing values with the text "NA" to ensure the DataFrame has no empty cells
```

```python
#Splitting the column and storing it in temp_df dataframe
temp_df = raw_df['TIMING'].str.split("(", n = 1, expand = True)
temp_df

#Word/Term              Definition

#temp_df              → New temporary DataFrame to store the split result
#raw_df['TIMING']     → Selects the TIMING column from the DataFrame
#.str                 → Accessor to apply string methods on a column
#.split("(")          → Splits each value at the "(" character
#n = 1                → Split only once (maximum 1 split per value)
#expand = True        → Expands the split result into separate columns (0 and 1)
#temp_df              → Displays the resulting temporary DataFrame

#Splits the TIMING column at "(" into two parts — the time text (col 0) and the extra info in brackets (col 1),stored in a new temp_df for further processing.
```

```python
#Assigning the columns back to the raw_df dataframe
raw_df['TIMING'] = temp_df[0]
raw_df['DAYS_OPEN'] = temp_df[1]
raw_df.head()

#Word/Term                Definition

#raw_df['TIMING']       → Selects/updates the TIMING column in raw_df
#raw_df['DAYS_OPEN']    → Creates a new column DAYS_OPEN in raw_df
#temp_df[0]             → Column 0 of temp_df — contains the time part
                         #(e.g. "11AM - 11PM")
#temp_df[1]             → Column 1 of temp_df — contains the days part
                         #(e.g. "Mon-Sun")
#=                      → Assigns the temp_df columns back to raw_df
#raw_df.head()          → Previews first 5 rows to verify the changes

#Assigns the split results back to raw_df — TIMING gets the clean time values and a new DAYS_OPEN column is created from the days/bracket part of the original TIMING column.
```

```python
#Removing the bracket character from Days column
raw_df['DAYS_OPEN'] = raw_df['DAYS_OPEN'].str.replace('\)','',regex=True)
raw_df.head()

#Word/Term                Definition

#raw_df['DAYS_OPEN']    → Selects the DAYS_OPEN column from raw_df
#.str                   → Accessor to apply string methods on the column
#.replace('\)', '')     → '\)' → escaped regex for literal ")" character
                         #''   → replaces it with empty string (removes it)
#regex=True             → Treats '\)' as a regex escaped pattern
#=                      → Reassigns the cleaned result back to DAYS_OPEN
#raw_df.head()          → Previews first 5 rows to verify the changes

#Removes the closing bracket ")" from DAYS_OPEN column using escaped regex '\)' so pandas treats it as a plain character(e.g. "Mon-Sun)" → "Mon-Sun").
```

```python
#Checking for Null records in DAYS_OPEN column
raw_df.isnull().sum()

#Word/Term        Definition

#raw_df         → Zomato Mumbai DataFrame
#.isnull()      → Returns True where value is missing (NaN),
                 #False where value exists
#.sum()         → Counts the number of True values per column
                 #(True=1, False=0)

#Checks if the newly created DAYS_OPEN column (or any othercolumn) has introduced any null values after the TIMING column split operation.
```

```python
#Replacing the Null values with 'NA'
raw_df.fillna('NA', inplace=True)

#Word/Term        Definition

#raw_df         → Zomato Mumbai DataFrame
#.fillna()      → Method that fills all missing (NaN) values
                 #with a specified replacement value
#'NA'           → Replacement text filled in place of NaN
#inplace=True   → Permanently modifies raw_df without reassignment

#Fills any new null values introduced in DAYS_OPEN (or other columns) after the TIMING split operation with 'NA', ensuring no empty cells remain in the DataFrame.
```

```python
#E.Removing the restaurant records whose Rating or Votes is 0
#Finding those restaurant whose has 0 Rating or Votes
useless_data = (raw_df['RATING'] == 0.0) | (raw_df['VOTES'] == 0)
raw_df[useless_data]

#Word/Term                    Definition

#useless_data               → Variable storing a True/False boolean mask
#raw_df['RATING'] == 0.0    → Returns True where RATING value is 0.0
#raw_df['VOTES'] == 0       → Returns True where VOTES value is 0
#|                          → OR operator — True if either condition is True
#(  )                       → Brackets group each condition separately
#useless_data               → Combined mask: True = restaurant has 0 Rating
                             #OR 0 Votes
#raw_df[useless_data]       → Filters and displays only those rows where
                             #useless_data is True

#Identifies restaurants with 0 Rating OR 0 Votes — these are newly listed or inactive restaurants with no useful data, so they need to be removed from the analysis.
```

```python
#Performing Negation of the useless dataset and then storing the correct data back in the raw_df DataFrame
#This permanently removes the wrong data from the original dataframe
raw_df = raw_df[~useless_data]

#Word/Term              Definition

#raw_df               → Zomato Mumbai DataFrame being updated
#=                    → Reassigns the filtered result back to raw_df
#raw_df[...]          → Filters rows from the DataFrame
#~                    → Negation operator — flips True→False, False→True
#useless_data         → Boolean mask where True = 0 Rating OR 0 Votes rows
#~useless_data        → Now True = valid restaurants, False = useless rows

#Permanently removes all restaurants with 0 Rating OR 0 Votes by keeping only the rows where ~useless_data is True, leaving only meaningful restaurant records for analysis.
```

```python
#Removing the irrelevant text from the Region column
raw_df['REGION'] = raw_df['REGION'].str.replace('[a-zA-Z].+-- ','',regex=True)

#Word/Term              Definition

#raw_df['REGION']     → Selects the REGION column from raw_df
#.str                 → Accessor to apply string methods on the column
#.replace()           → Substitutes matched pattern with a new value
#'[a-zA-Z].+-- '     → Regex pattern:
#   [a-zA-Z]         →   Matches any letter (upper or lowercase)
#   .+               →   Followed by one or more of any character
#   --               →   Followed by literal "--"
#   ' '              →   Followed by a space
#''                   → Replaces matched text with empty string (removes it)
#regex=True           → Treats the pattern as a regex expression
#=                    → Reassigns the cleaned result back to REGION

#Removes irrelevant prefix text (e.g. "Restaurants in Mumbai -- ")from the REGION column, keeping only the clean locality/area name (e.g. "Bandra", "Andheri").
```

```python
#Removing the West & East from the Region column
raw_df['REGION'] = raw_df['REGION'].str.replace(' West| west| East| east','',regex=True)

#Word/Term                    Definition

#raw_df['REGION']           → Selects the REGION column from raw_df
#.str                       → Accessor to apply string methods on the column
#.replace()                 → Substitutes matched pattern with a new value
#' West| west| East| east'  → Regex pattern:
#   ' West'                 →   Matches " West" (with space before)
#   | west                  →   OR " west" (lowercase)
#   | East                  →   OR " East"
#   | east                  →   OR " east" (lowercase)
#   |                       →   OR operator in regex
#''                         → Replaces all matched text with empty string
#regex=True                 → Treats the pattern as a regex expression
#=                          → Reassigns the cleaned result back to REGION

#Removes directional suffixes "West" and "East" (both upper and lowercase) from REGION values to standardize area names (e.g. "Andheri West" → "Andheri").
```

```python
#Q10)Find the list of Affordable Restaurants?
```

The criteria for Affordable Restaurants would be:-
1) Low Price 2) High Rated
First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.
Let me explain:-The most expensive restaurant has an average meal cost= 6000. We'll try to stay economical and only pick the
restaurants that are 1/4th of 6000

```python
max_price = zomato_df['PRICE'].max()
one_fourth_price = max_price/4
one_fourth_price

#Word/Term                  Definition

#max_price                → Variable storing the highest price value
#zomato_df['PRICE']       → Selects the PRICE column from zomato_df
#.max()                   → Returns the maximum (highest) value in PRICE
#one_fourth_price          → Variable storing 1/4th of the max price
#max_price / 4              → Divides max_price by 4
#one_fourth_price          → Displays the calculated one-fourth value

#Calculates the maximum restaurant price and finds its one-fourth (1/4th) value, used as a threshold to identify expensive restaurants that are 1/4th of the max price.
```

```python
#Finding list of restaurants that have price less than and equal to 1/4th of the max price i.e Finding Cheap Restaurant
aff_rest_df = zomato_df[['NAME', 'PRICE', 'CUSINE_CATEGORY', 'REGION', 'CUSINE TYPE']]
aff_rest_df = aff_rest_df[aff_rest_df['PRICE'] <= 1250]
aff_rest_df.sort_values(by='PRICE', inplace=True)
aff_rest_df

#Word/Term                              Definition

#aff_rest_df                          → New DataFrame to store affordable
                                        #(cheap) restaurants
#zomato_df[[...]]                     → Selects only specific columns:
#['NAME', 'PRICE',                      NAME, PRICE, CUSINE_CATEGORY,
# 'CUSINE_CATEGORY', 'REGION',          REGION, CUSINE TYPE
# 'CUSINE TYPE']
#aff_rest_df['PRICE'] <= 1250         → Filters rows where PRICE is less than
                                        #or equal to 1250 (1/4th of max price)
#aff_rest_df[...]                     → Keeps only affordable restaurant rows
#.sort_values()                       → Sorts the DataFrame by a specified column
#by='PRICE'                           → Sorts based on the PRICE column
#inplace=True                         → Permanently modifies aff_rest_df
                                        #without reassignment
#aff_rest_df                          → Displays the final sorted DataFrame

#Filters and sorts restaurants priced at or below 1250 (1/4th of the max price), identifying affordable/cheap restaurant options across Mumbai.
```

```python
#Finding the highest rated list of restaurants
highrate_rest_df = zomato_df[['NAME', 'PRICE', 'CUSINE_CATEGORY', 'REGION', 'CUSINE TYPE','RATING']]
highrate_rest_df = highrate_rest_df[highrate_rest_df['RATING'] >= 4.5]
highrate_rest_df.sort_values(by='PRICE', inplace=True)
highrate_rest_df

#Word/Term                              Definition

#highrate_rest_df                     → New DataFrame to store highest
                                        #rated restaurants
#zomato_df[[...]]                     → Selects only specific columns:
#['NAME', 'PRICE',                      NAME, PRICE, CUSINE_CATEGORY,
# 'CUSINE_CATEGORY', 'REGION',          REGION, CUSINE TYPE, RATING
# 'CUSINE TYPE', 'RATING']
#highrate_rest_df['RATING'] >= 4.5    → Filters rows where RATING is
                                        #4.5 or higher
#highrate_rest_df[...]                → Keeps only highest rated restaurant rows
#.sort_values()                       → Sorts the DataFrame by a specified column
#by='PRICE'                           → Sorts based on the PRICE column
#inplace=True                         → Permanently modifies highrate_rest_df
                                        #without reassignment
#highrate_rest_df                     → Displays the final sorted DataFrame

#Filters and sorts restaurants with a rating of 4.5 or above, creating a list of the highest rated restaurants in Mumbai ordered by price.
```

Now, we'll merge the aff_rest_df with highrate_rest_df to obtain the intersection i.e the list of Affordable Restaurants !!

```python
highrate_aff_df = pd.merge(aff_rest_df, highrate_rest_df, how='inner', on=['NAME', 'REGION'])
highrate_aff_df = highrate_aff_df[['NAME', 'PRICE_x', 'CUSINE_CATEGORY_x', 'REGION', 'CUSINE TYPE_x']]
highrate_aff_df.rename(columns={'NAME':'NAME', 'PRICE_x':'PRICE', 'CUSINE_CATEGORY_x':'CUSINE_CATEGORY',
'REGION':'REGION', 'CUSINE TYPE_x':'CUSINE TYPE'},inplace=True)

#Word/Term                              Definition

#highrate_aff_df                      → New DataFrame storing restaurants that
                                        #are BOTH affordable AND highest rated
#pd.merge()                           → Combines two DataFrames based on
                                        #matching column values
#aff_rest_df                          → Affordable restaurants DataFrame
#highrate_rest_df                     → Highest rated restaurants DataFrame
#how='inner'                          → Keeps only rows that match in BOTH
                                        #DataFrames (intersection)
#on=['NAME', 'REGION']                → Matches rows where NAME and REGION
                                        #are the same in both DataFrames

#highrate_aff_df[[...]]               → Selects specific columns after merge
#'PRICE_x', 'CUSINE_CATEGORY_x',      → "_x" suffix added automatically since
#'CUSINE TYPE_x'                        these columns existed in both DataFrames
                                        #(keeping the version from aff_rest_df)

#.rename(columns={...})               → Renames the "_x" suffixed columns back
                                        #to their original clean names
#inplace=True                         → Permanently modifies highrate_aff_df

#Merges affordable restaurants with highest rated restaurants to find the best of both — restaurants that are cheap (≤1250 price) AND highly rated (≥4.5 rating).
```
